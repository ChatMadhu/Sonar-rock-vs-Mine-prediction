Import the libraries

```python
import numpy as np                                    #this is for creating numpy arrays
import pandas as pd                                   #this is for data loading into pandas data frame
from sklearn.model_selection import train_test_split  #this is to devide the data set into two as training and testing data sets
from sklearn.linear_model import LogisticRegression   #this is to select the algorithem, which is used in this project
from sklearn.metrics import accuracy_score            #this is to find the accuracy of the created model
```

Data collection and data processing

```python
sonar_data = pd.read_csv('/content/sonar data.csv', header = None)    #first we have to load our csv data set into pandas dataframe.
                                                                      #If you check you can find that there is no headers in the dcsv data set. so it have to be mentioned for this pandas data frame also.
```

```python
sonar_data.groupby(60).mean()     #by this we're going to find the mean value for each row and using that value we are going to predict the rock or minne
```

```python
# Lets seperate the data and lables. This is a supervised learning design. So we're labeling the values
#it means the data and the target have to be seperated. The data will be labeled as 0 to 207 and the target is to be labeled as M and R.

X = sonar_data.drop(columns=60,  axis=1)
Y = sonar_data[60]
print(X)
print(Y)
```

Lets seperate the data set as training and testing data. 

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify = Y, random_state = 1)

# here the X and Y is the training and testing data. And 0.1 test size is that we are using 10% of the data set for the training. the stratify means that we need to split the data according to  the y values. it  means for the testing data and also for trainig data, the y parameters should be spreaded evenly.
#random state means the splitting style of data by the algorithem. if we split the data as random_state=1, it splits data in one way. For random state=2 it splits the data in another style and give another portion of data as training and also for testing.
```

```python
print(X.shape, X_test.shape, X_train.shape)
print(X_train)
print(Y_train)
```

Model Training ------> Logistic Regression

```python
#now lets train the data with logistic regression model

model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
# now the model should be evaluated

# 1) for that first we should find the accuracy on testing data. If the accuracy is more than 70% it is good

X_train_prediction = model.predict(X_train)                           #X train prediction is the prediction of our model.
training_data_accuracy = accuracy_score(X_train_prediction, Y_train) #Y train is the actual answer of the predicting data. from this accuracy score, the predicted values and the actual values are compared each and give an accuracy for that results.
print(training_data_accuracy, '= Training data Accuracy')
```

```python
# now we will see how much is the testing data accuracy.
X_test_prediction = model.predict(X_test)                           #X train prediction is the prediction of our model.
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test) #Y train is the actual answer of the predicting data. from this accuracy score, the predicted values and the actual values are compared each and give an accuracy for that results.
print(testing_data_accuracy, '= Testing data Accuracy')
```
